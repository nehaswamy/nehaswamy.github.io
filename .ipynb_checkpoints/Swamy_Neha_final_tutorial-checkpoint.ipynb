```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
import numpy as np 
import matplotlib
from matplotlib import pyplot
import sys
import seaborn as sns
import re
!{sys.executable} -m pip install lxml
```

```python
def get_table(url, x):
    r = requests.get(url)
    text = r.text
    soup = BeautifulSoup(text, 'html.parser')
    table = soup.find_all('table')[x]
    
    return table
```

# Physical Album Sales

## Best-Selling Albums in South Korea by Year

```python
url = "https://en.wikipedia.org/wiki/List_of_best-selling_albums_in_South_Korea"
table = get_table(url, 3)
columns = ['Year', 'Artist', 'Album', 'Sales', 'Ref']
df = pd.read_html(table.prettify())[0]
df.columns = columns
df = df.drop(labels = ['Ref'], axis = 1)
df = df.sort_values(by = ['Year'])
df.index = range(len(df))

for index, row in df.iterrows():
    df.at[index, 'Year'] = int(str(df.at[index, 'Year'])[0:4])

pyplot.figure(figsize = (20, 10))
plot = sns.barplot(x = 'Year', y = 'Sales', hue = 'Artist', data = df, dodge = False, palette = sns.color_palette('Paired'))

pyplot.ylabel('Sales (in Millions)')
pyplot.xlabel('Year')
pyplot.title('Best Selling Albums in South Korea by Year (1998 - 2019)')
pyplot.show()
```

From the above bar graph, we can see that EXO sold the most albums for five consecutive years (2013 - 2017). However, BTS took over EXO as the best-selling group in 2018 and 2019. In addition to EXO and BTS, there are some artists who have been the best-selling artist for more than one year, including 
* Seo Taiji (1998, 2004)
* Jo Sungmo (1999, 2000)
* SG Wannabe (2005, 2007)
* TVXQ (2006, 2008)
* Super Junior (2009, 2011, 2012)

# Digital Sales

## Best-Selling Singles in South Korea by Year

```python
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_singles_in_South_Korea'
table = get_table(url, 1)

columns = ['Year', 'Song', 'Artist', 'Yearly Sales', 'Total Sales']
df = pd.read_html(table.prettify())[0]
df.columns = columns
df = df.sort_values(by = ['Year'])
df.index = range(len(df))
df = df.drop(index = len(df) - 1)

for index, row in df.iterrows():
    df.at[index, 'Yearly Sales'] = df.at[index, 'Yearly Sales'].replace(',', '')
    df.at[index, 'Total Sales'] = df.at[index, 'Total Sales'].replace(',', '')
    
    pattern = '\d+'
    df.at[index, 'Yearly Sales'] = re.findall(pattern, df.at[index, 'Yearly Sales'])[0]
    df.at[index, 'Total Sales'] = re.findall(pattern, df.at[index, 'Total Sales'])[0]

pyplot.figure(figsize = (20, 10))
plot = sns.barplot(x = 'Year', y = 'Yearly Sales', hue = 'Artist', data = df, dodge = False, palette = sns.color_palette('Paired'))

pyplot.title('Best-Selling Singles in South Korea by Year (2010 - 2018)')
pyplot.show()
```

Here, we can see that the artists with the best-selling digital singles are completely different from the artists who have sold the most physical albums. We do not see Girls' Generation, Super Junior, EXO, or BTS in this graph, even though they sold the most physical albums from 2010 to 2019. Thus, we can see that different artists dominate different areas. 

However, let's see if there is an association between digital single sales and physical sales. We will use the following artists from the best-selling physical album sale data:
* TVXQ 
* Super Junior
* Girls' Generation 
* EXO
* BTS 

```python
def get_yearly_downloads(df):
    years = df['Year'].drop_duplicates() 
    sales = pd.DataFrame(columns = ['Year', 'Downloads'])
    index = 0

    for year in years:
        s = 0
        downloads = df[df.Year == year]['Downloads']
    
        for download in downloads:
            s += download
        
        sales.at[index, 'Year'] = year
        sales.at[index, 'Downloads'] = s
        index += 1
    
    return sales
```

```python
url = "https://en.wikipedia.org/wiki/Exo_discography"

table = get_table(url, 5)
exo = pd.read_html(table.prettify())[0]
exo.columns = ['Title', 'Year', 'Peak (Gaon)', 'Peak (Billboard)', 'Peak (Chinese)', 'Peak (US World)', 'Downloads', 'Album']

exo = exo.drop(index = len(exo) - 1)
exo = exo.drop(index = 159)

for index, row in exo.iterrows():
    for column in exo.columns:
        if exo.at[index, column] == "—":
            exo.at[index, column] = np.nan
    
    pattern = "\d+,\d+"
    downloads = str(exo.at[index, 'Downloads'])
    match = re.findall(pattern, downloads)
    
    if (match != []):
        exo.at[index, 'Downloads'] = int(match[0].replace(',', ''))
    
    else:
        exo.at[index, 'Downloads'] = 0
    
downloads1 = get_yearly_downloads(exo)

table = get_table(url, 4)
exo = pd.read_html(table.prettify())[0]
exo.columns = ['Title', 'Year', 'Gaon', 'Billboard', 'CHN', 'Oricon', 'Hot 100', 'CAN', 'FRA Dig', 'MAL', 'SIN', 'US World', 'Downloads', 'Certifications', 'Album', 'Unknown']
exo = exo.drop(index = 29)
exo = exo.drop(index = 24)
exo = exo.drop(labels = 'Unknown', axis = 1)

for index, row in exo.iterrows():
    for column in exo.columns:
        if exo.at[index, column] == "—":
            exo.at[index, column] = np.nan
        
    if (exo.at[index, 'Downloads'] is np.nan):
        exo.at[index, 'Downloads'] = 0
    
    pattern = "\d+,\d+"
    downloads = str(exo.at[index, 'Downloads'])
    matches = re.findall(pattern, downloads)
    s = 0
    
    for match in matches:
        s += int(match.replace(',', ''))
    
    exo.at[index, 'Downloads'] = s

downloads2 = get_yearly_downloads(exo)
index = 0

for index, row in downloads1.iterrows():
    dl1 = downloads1.at[index, 'Downloads']
    dl2 = downloads2.at[index, 'Downloads']
    
    downloads1.at[index, 'Downloads'] = dl1 + dl2
    index += 1
```
