## Import modules

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np 
from datetime import datetime
import re
import sys
import matplotlib as mpl
!{sys.executable} -m pip install lxml
```

## Step 1

```python
# Website with data to be scraped
url = "https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares"

# Header to be used in the request
header = {"user-agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36"}

# GET the webpage and find the table on the page
r = requests.get(url, headers = header)
text = r.text
soup = BeautifulSoup(text, 'html.parser')
table = soup.find('table')

# Create a dataframe with the following columns and data from the table 
columns = ['rank', 'x_classification', 'date', 'region', 'start_time', 'maximum_time', 'end_time', 'movie']
df = pd.read_html(table.prettify())[0]
df.columns = columns
df.index = range(1, 51)

df
```

This section of code uses the requests library to get the Top 50 Solar Flares article from SpaceWeatherLive.com and the BeautifulSoup library to pull the data from the table on the page. The data is then stored in a DataFrame. 

## Step 2

```python
# Remove the 'movie' column
df = df.drop(labels = 'movie', axis=1)

'''
Go through each row and convert the start, maximum, and end times into datetime objects. Update any missing regions
with NaN
'''
for row in df.iterrows():
    date = row[1]['date']
    start_time = date + ' ' + row[1]['start_time']
    maximum_time = date + ' ' + row[1]['maximum_time']
    end_time = date + ' ' + row[1]['end_time']
    region = row[1]['region']
    
    st = datetime.strptime(start_time, "%Y/%m/%d %H:%M")
    mt = datetime.strptime(maximum_time, "%Y/%m/%d %H:%M")
    et = datetime.strptime(end_time, "%Y/%m/%d %H:%M")
    
    df.at[row[0], 'start_time'] = st
    df.at[row[0], 'maximum_time'] = mt
    df.at[row[0], 'end_time'] = et

    if (region == '-'):
        df.at[row[0], 'region'] = np.nan
    
# Remove the date column and update the remaining column names
df = df.drop(labels = 'date', axis = 1)
columns = ['rank', 'x_classification', 'region', 'start_datetime', 'maximum_datetime', 'end_datetime']
df.columns = columns

df
```

For part 2, the movie column is removed from the DataFrame. The start, maximum, and end times are then converted to datetime objects. Finally, missing regions are encoded as missing (NaN). 

## Step 3

```python
# Get the data from the NASA web page and put the data from lines 15 to 533 into a DataFrame
url = 'http://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html'
r = requests.get(url)
text = r.text
soup = BeautifulSoup(text, 'html.parser')
data = soup.get_text()
start = 15
end = 533
data = data.split("\n")[start:end]

# Only store the data in the first 15 columns
for i in range(0, len(data)):
    data[i] = data[i].split()[0:15]

# Store the data into a DataFrame and add appropriate column names and indexes
df2 = pd.DataFrame(data)
columns = ["start_date", "start_time", "end_date", "end_time", "start_frequency", "end_frequency", "flare_location", 
           "flare_region", "flare_classification", "cme_date", "cme_time", "cme_angle", 
           "cme_width", "cme_speed", "plot"]
df2.columns = columns
df2.index = range(1, len(data) + 1)

df2
```

This section gets the CME_list article from the NASA website using the requests library and pulls the data using the BeautifulSoup library. The data from the website is then restructured such that any titles, headers, descriptions, etc. are removed and only the 15 relevant columns of the table remain.

## Step 4

```python
# Used to update the datetime objects for sub-problem 4 
def update_date_time(dt):
    udt = datetime(dt.year, dt.month, dt.day, hour = dt.hour, minute = dt.minute, second = dt.second)
    day = dt.day + 1
    thirty_day_months = [4, 6, 9, 11]

    if (day == 29 and dt.month == 2 and dt.year % 4 != 0):
        udt = dt.replace(day = 1, month = 3)
  
    elif (day == 31 and dt.month in thirty_day_months):
        month = dt.month + 1
        udt = dt.replace(day = 1, month = month)
  
    elif (day == 32):
        month = dt.month + 1

        if (month == 13):
            udt = dt.replace(day = 1, month = 1, year = dt.year + 1)
    
        else:
            udt = dt.replace(day = 1, month = month)
            
    else:
        udt = dt.replace(day = day)
        
    return udt

# Updated column names of the new DataFrame
columns = ["startdate_time", "enddate_time", "start_frequency", "end_frequency", "flare_location", "flare_region", 
           "importance", "cmedate_time", "cpa", "width", "speed", "plot", "is_halo", 
           "width_lower_bound"]

# Add two new columns 
df2.insert(len(df2.columns), 'is_halo', "")
df2.insert(len(df2.columns), 'width_lower_bound', "")

# Iterate through the rows of the DataFrame
for row in df2.iterrows():
    # 1. Recode any missing entries as NaN
    for entry in df2.columns:
        value = str(row[1][entry])
        if ('-' in value or '?' in value):
            df2.at[row[0], entry] = np.nan

    # 2. Update cme_angle and is_halo columns
    cpa = row[1]['cme_angle']
    halo = row[1]['is_halo']
    
    if ('Halo' == cpa):
        df2.at[row[0], 'cme_angle'] = np.nan
        df2.at[row[0], 'is_halo'] = True
    
    else:
        df2.at[row[0], 'is_halo'] = False
    
    # 3. Update width_lower_bound column
    width = row[1]['cme_width']
    width_lower_bound = row[1]['width_lower_bound']
    
    if ('>' in str(width)):
        df2.at[row[0], 'width_lower_bound'] = True
    
    else:
        df2.at[row[0], 'width_lower_bound'] = False
            
    # 4. Combine date and time columns for start, end, and cme so they can be endocoded as datetime objects
    
    # Used to check if the time is 24:00 for any of the dates
    is_next_day_start = False
    is_next_day_end = False 
    is_next_day_cme = False 
    
    # Used to check if each of the dates are in a valid format
    match = '[0-9]{4}/[0-9]{1,2}/[0-9]{1,2}'
    
    # Get the year since only the start_date column includes the year
    year = row[1]['start_date'].split("/")[0]
    
    # Create strings for each of the dates
    start_date = row[1]['start_date']
    end_date = year + "/" + str(row[1]['end_date'])
    cme_date = year + "/" + str(row[1]['cme_date'])
    
    # Get the start, end, and cme times
    start_time = str(row[1]['start_time'])
    end_time = str(row[1]['end_time'])
    cme_time = str(row[1]['cme_time'])
    
    # Check if the time is 24:00 for each of the times
    if ('24' == start_time[0:2]):
        minute = start_time.split(":")[1]
        start_time = "00" + ":" + minute 
        is_next_day_start = True
    
    if ('24' == end_time[0:2]):
        minute = end_time.split(":")[1]
        end_time = "00" + ":" + minute
        is_next_day_end = True
    
    if ('24' == cme_time[0:2]):
        minute = cme_time.split(":")[1]
        cme_time = "00" + ":" + minute
        is_next_day_cme = True
    
    # Combine the dates and times
    start_time = start_date + ' ' + start_time 
    end_time = end_date + ' ' + end_time 
    cme_time = cme_date + ' ' + cme_time
    
    # datetime objects that are initially NaN 
    st = np.nan
    et = np.nan
    cmet = np.nan
    
    '''
    See if start_time, end_time, and cme_time match the format. If they do, then see if the time is for the next
    day. If so, then update the day, month, and year for st, et, and cmet accordingly. 
    '''
    if (re.match(match, start_time)):
        st = datetime.strptime(start_time, "%Y/%m/%d %H:%M")
        
        if (is_next_day_start):
            st = update_date_time(st)
        
    if (re.match(match, end_time)):
        et = datetime.strptime(end_time, "%Y/%m/%d %H:%M")
        
        if (is_next_day_end):
            et = update_date_time(et)
        
    if (re.match(match, cme_time)):
        cmet = datetime.strptime(cme_time, "%Y/%m/%d %H:%M")
        
        if (is_next_day_cme):
            cmet = update_date_time(cmet)
    
    # Replace the start_time, end_time, and cme_time columns with st, et, and cmet, respectively
    df2.at[row[0], 'start_time'] = st
    df2.at[row[0], 'end_time'] = et
    df2.at[row[0], 'cme_time'] = cmet

# Update the columns of df2
df2 = df2.drop(['start_date', 'end_date', 'cme_date'], axis=1)
df2.columns = columns

df2
```

Two new columns, is_halo and width_lower_bound, are added to the DataFrame. is_halo is True for each row that has the string 'Halo' in the cpa column and False otherwise. width_lower_bound is True for each row that has a ">" in the width column and False otherwise. The start, cme, and end time columns are then converted to datetime objects.

I sorted the data by first sorting the DataFrame by the classification column and getting all the rows with "X" in the classification column. I then extracted the number after the X and sorted the column by the numerical value since the data in the SpaceWeatherLive.com table is sorted by the number after the X. Next, I got the top 50 rows by classification and added the X back to the classification for each row. 

I was able to replicate the data from SpaceWeatherLive.com quite well but not perfectly. The NASA data does not include a max_date and a max_time column, does not include all the flares in terms of classification, and has different times for some of the classifications that are the same. 

## Question 2

```python
def find_best_matching_row(df, row):
    # Best matching row to be returned
    best_match = ''
    
    # Get the values in the columns of row
    xc = str(row[1]['x_classification'])
    region = str(row[1]['region'])
    sdt = str(row[1]['start_datetime'])
    edt = str(row[1]['end_datetime'])
    start_date = str(row[1]['start_datetime']).split(' ')[0]
    end_date = str(row[1]['end_datetime']).split(' ')[0]
    rank = str(row[1]['rank'])
    
    # Remove the + in the x_classification column if it is there
    if ('+' in xc):
        xc = xc.strip('+')
    
    # Iterate through the rows in the DataFrame to find a match
    for r in df.iterrows():
        # Get the values of the relevant columns in the DataFrame
        df_xc = str(r[1]['importance'])
        df_region = str(r[1]['flare_region'])
        df_sdt = str(r[1]['startdate_time'])
        df_edt = str(r[1]['enddate_time'])
        df_rank = str(r[1]['SpaceWeatherLive_rank'])
        
        # Match the row based on priority (see the cell below for an explanation)
        if (df_xc[len(df_xc) - 1] == '.'):
            df_xc = df_xc.strip('.')
        
        if (df_xc == xc and df_region == region and df_sdt == sdt and df_edt == edt and df_rank == ''):
            best_match = r 
            break; 
        
        elif (df_xc == xc and df_region == region and start_date in sdt and end_date in edt and df_rank == ''):
            best_match = r 
            break; 
        
        elif (df_xc == xc and df_sdt == sdt and df_edt == edt and df_rank == ''):
            best_match = r 
            break; 
        
        elif (df_xc == xc and (df_sdt == sdt or df_edt == edt) and df_rank == ''):
            best_match = r 
            break;
        
        elif (df_xc == xc and start_date in sdt and end_date in edt and df_rank == ''): 
            best_match = r 
            break;
        
        elif (df_xc == xc and (start_date in sdt or end_date in edt) and df_rank == ''):
            best_match = r 
            break;
        
        elif (df_sdt == sdt and df_edt == edt and df_rank == ''):
            best_match = r
        
        elif ((df_sdt == sdt or df_edt == edt) and df_rank == ''):
            best_match = r; 
        
        elif (start_date in sdt and end_date in edt and df_rank == ''):
            best_match = r 
        
        elif ((start_date in sdt or end_date in edt) and df_rank == ''):
            best_match = r; 

    # Return the best matching row or '' if there was no matching row
    return best_match 

# Insert a column representing the corresponding rank in the SpaceWeatherLive DataFrame (df) into the NASA DataFrame(df2)
df2.insert(len(df2.columns), 'SpaceWeatherLive_rank', '')

# Iterate through the rows of the SpaceWeatherLive DataFrame to find the matching row in df2
for row in df.iterrows():
    best_match = find_best_matching_row(df2, row)[0]
    rank = row[1]['rank']
    
    if (best_match != ''):
        df2.at[best_match, 'SpaceWeatherLive_rank'] = rank 

# If there was no match, code the value in the SpaceWeatherLive_rank with NaN for each row in df2
for row in df2.iterrows():
    if (row[1]['SpaceWeatherLive_rank'] == ''):
        df2.at[row[0], 'SpaceWeatherLive_rank'] = np.nan
        
df2
```

To pick the best matching row, I matched each row in order of priority:

1. The x_classification/importance, region, start_datetime, and end_datetime are the same
2. The x_classification/importance, region, start date, and end date (but not the start and end times) are the same
3. The x_classification/importance, start_datetime, and end_datetime are the same
4. The x_classification/importance and either the start_datetime or end_datetime are the same 
5. The x_classification/importance, start date, and end date (but not the start and end time) are the same
6. The x_classification/importance and either the start date or end date are the same
7. The start_datetime and end_datetime are the same 
8. Either the start_datetime or the end_datetime is the same 
9. The start date and end date (but not the start and end time) are the same 
10. Either the start date or the end date (but not the start and end time) is the same

I also checked to see if the row in df2 does not already have a rank to prevent overriding.
