```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
import numpy as np 
import matplotlib
from matplotlib import pyplot
import sys
import seaborn as sns
import re
!{sys.executable} -m pip install lxml
```

```python
def get_dataframe(url, table_number):
    r = requests.get(url)
    text = r.text
    soup = BeautifulSoup(text, 'html.parser')
    table = soup.find_all('table')[table_number]
    df = pd.read_html(table.prettify())[0]
    
    return df

def get_sales(df, digital_only = False):
    for index, row in df.iterrows():
        sales = str(df.at[index, 'Sales'])
        
        if (sales != 'nan'):
            sales = sales.replace(',', '')
            sales = re.sub('\[[A-Z0-9]+\]', '', sales)
            s = 0
        
            if (digital_only == False):
                matches = re.findall("\d+", sales)
    
                for match in matches:
                    s += int(match)
    
            else:        
                matches = re.findall("(\d+)(\s*\(\s*Phy.\s*|\s*Dig.\s*\))?", sales)
                
                for match in matches:
                    if ('Phy.' not in str(match[1])):
                        s += int(match[0])
                
            if (s == 0):
                s = np.nan
            
            df.at[index, 'Sales'] = s
    
    return df

def get_years(df):
    for index, row in df.iterrows():
        year = re.findall('\d{4}', df.at[index, 'Details'])[0]
        df.at[index, 'Year'] = int(year)
    
    return df

def get_pretty_dataframe(df, digital_only = False):  
    if ('Year' not in df.columns):
        df.insert(2, 'Year', 0)
        df = get_years(df)
        
    df = get_sales(df, digital_only)
    
    if ('Details' in df.columns):
        df = df.drop(labels = ['Details'], axis = 1)
    
    if ('' in df.columns):
        df = df.drop(labels = [''], axis = 1)
    
    for index, row in df.iterrows():
        for column in df.columns:
            value = str(df.at[index, column])
            
            if (value == "" or value == "nan" or "—" in value or '*' in value):
                df.at[index, column] = np.nan
            
            else:
                value = value.replace(',', '')
                value = re.sub('\[[A-Za-z0-9]+\]', '', value)
            
                matches = re.findall('\d+', value)
            
                if (len(matches) > 0 and value == matches[0]):
                    value = int(value)
            
                df.at[index, column] = value

        df.at[index, 'Title'] = df.at[index, 'Title'].replace('"', '').strip()
            
    df = df.sort_values(by = ['Year'])
    df.index = range(1, len(df) + 1)
    
    return df
```

## 3. Exploratory Data Analysis

### 3.1: Girls' Generation vs. Twice: Physical and Digital Sales

```python
def plot_total_sales(df1, df2, group1, group2, length1, length2):
    sales1 = df1.groupby(by = ['Year'])['Sales'].sum()
    sales2 = df2.groupby(by = ['Year'])['Sales'].sum()

    plot = pyplot.figure(figsize = (10, 5))
    plot = pyplot.plot(range(length1), np.asarray(sales1))
    plot = pyplot.plot(range(length2), np.asarray(sales2))
    plot = pyplot.legend([group1, group2])
    plot = pyplot.xlabel('Years Since Debut')
    plot = pyplot.ylabel('Sales')
    
    return plot
```

```python
def plot_average_sales(df1, df2, group1, group2, debut1, debut2, length):
    df1['Sales'] = df1['Sales'].astype('float')
    df2['Sales'] = df2['Sales'].astype('float')
    
    avg_sales1 = df1.groupby(by = ['Year'])['Sales'].mean().fillna(0)
    avg_sales2 = df2.groupby(by = ['Year'])['Sales'].mean().fillna(0)
    
    plot = pyplot.figure(figsize = (10, 5))
    plot = pyplot.plot(range(len(avg_sales1)), avg_sales1)
    plot = pyplot.plot(range(len(avg_sales2)), avg_sales2)
    plot = pyplot.xlabel('Years Since Debut')
    plot = pyplot.ylabel('Average Sales')
    plot = pyplot.legend([group1, group2])
    
    return plot
```

```python
plot = plot_total_sales(snsd_albums, twice_albums, "Girls' Generation", "Twice", 9, 6)
pyplot.show()
```

```python
plot = plot_average_sales(snsd_albums, twice_albums, "Girls' Generation", "Twice", 2007, 2015, 11)
pyplot.show()
```

### 3.2: Girls' Generation vs. Twice: Number of Singles that Peaked in the Top 50

```python
def count_top_50(df, columns, snsd = False):
    count = 0
    
    for index, row in df.iterrows():
        exists = False
        
        for column in columns:
            value = df.at[index, column]
            
            if (value is not np.nan and str(value) != 'To be released' and int(value) <= 50):
                if (snsd == True):
                    if (df.at[index, 'Year'] <= 2012):
                        exists = True
                
                else:
                    exists = True
    
        if (exists):
            count += 1
    
    return count
```

```python
columns = ['Gaon', 'KOR Hot', 'Oricon', 'JPN Hot', 'RIAJ', 'TW', 'US World']
snsd_top_50 = count_top_50(snsd_digitals, columns, snsd = True)

columns = ['KOR', 'KOR Hot', 'CAN', 'JPN', 'JPN HOT', 'JPN Dig.', 'NZ Hot', 'US World']
twice_top_50 = count_top_50(twice_digitals, columns)

pyplot.figure(figsize = (10, 5))
pyplot.bar(x = ["Girls' Generation", "Twice"], height = [snsd_top_50, twice_top_50], color = 'pink')
pyplot.xlabel('Group')
pyplot.ylabel('Sales')
pyplot.title("Number of Songs that Peaked in the Top 50 (0 - 5 Years After Debut)")
pyplot.show()
```

```python
snsd_count = snsd_top_50 / len(snsd_digitals[snsd_digitals.Year <= 2012])
twice_count = twice_top_50 / len(twice_digitals)

pyplot.figure(figsize = (10, 5))
pyplot.bar(x = ["Girls' Generation", "Twice"], height = [snsd_count, twice_count], color = 'pink')
pyplot.xlabel('Group')
pyplot.ylabel('Sales')
pyplot.title("Proportion of Songs that Peaked in the Top 50 (0 - 5 Years After Debut)")
pyplot.show()
```
